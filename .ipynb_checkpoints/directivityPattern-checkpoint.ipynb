```python
import numpy as np
import scipy.io.wavfile as wav
import scipy.fft
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Define constants and initialize variables
step = 5
angles = np.arange(0, 360, step)
num_angles = len(angles)
ir_data = [None] * num_angles
fft_data = [None] * num_angles

polar_freq = [500, 1000, 2000, 4000, 8000]
num_polar_freq = len(polar_freq)
polars = np.zeros((num_polar_freq, num_angles))
Zs = np.zeros((num_polar_freq, num_angles))

# Read audio files and compute FFT
channel = 1
plt.figure()
for i in range(num_angles):
    filename = f"AnechoicRoomMeasurements/IRs_Channel_{channel}_Hann/IR_monKD8N255G_{angles[i]}_Channel_{channel}.wav"
    fs, ir_data[i] = wav.read(filename)
    N = len(ir_data[i])
    freq = np.fft.fftfreq(N, 1/fs)[:N//2]
    fft_data[i] = np.fft.fft(ir_data[i])
    Z = fft_data[i][:N//2]
    magnitude = 20 * np.log10(np.abs(Z))
    for j in range(num_polar_freq):
        index = (i + int(270 / step)) % int(360 / step)
        freq_index = int(polar_freq[j] * N / fs)
        polars[j, index] = magnitude[freq_index]
        Zs[j, index] = Z[freq_index]
    plt.plot(freq, magnitude)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude (dB)")
plt.title(f"FFTs for Channel {channel}")
plt.show()

# Plot directivity pattern
plt.figure()
angles = np.append(angles, 360)
polars = np.hstack((polars, polars[:, [0]]))
Zs = np.hstack((Zs, Zs[:, [0]]))
maxMag = np.max(polars)
polars = polars - maxMag

tbl = pd.DataFrame(np.column_stack((np.deg2rad(angles), polars.T)),
                   columns=["Angles (rad)"] + [f"{freq}Hz" for freq in polar_freq])

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
for col in tbl.columns[1:]:
    ax.plot(tbl["Angles (rad)"], tbl[col], label=col, linewidth=1)

ax.set_thetalim(0, 2 * np.pi)
ax.set_thetagrids(range(0, 360, 45))
ax.set_rlim(-60, 0)
ax.set_rticks(range(-60, 5, 5))
plt.title(f"Directivity Pattern for Channel {channel}")
plt.legend()
plt.show()
```

```python
# Save the data
np.savez("MatData/polars.npz", polars=polars, polar_freq=polar_freq)
```
